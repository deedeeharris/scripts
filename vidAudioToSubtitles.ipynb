```python
import youtube_dl

def download_audio(url):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': '%(title)s.%(ext)s',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }]
    }
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])
```

```python
# @title Run me to setup

# declare paths to source vid
import moviepy.editor as mp
import os

# function will read video from path, and save extracted audio as mp3
def extract_audio(video_source_path, audio_dest_path):
  my_clip = mp.VideoFileClip(video_source_path) # path to your vid
  my_clip.audio.write_audiofile(audio_dest_path) # path to your extracted audio file
```

## Increase audio volume (only if needed)

```python
from pydub import AudioSegment

def increase_volume(input_file, output_file, gain):
    # Read in the audio file
    audio = AudioSegment.from_file(input_file)

    # Increase the volume
    audio = audio.apply_gain(gain)

    # Save the modified audio data to a new file
    audio.export(output_file, format=input_file.split('.')[-1])
```

```python
# load the model (make sure you are on GPU runtime)
from transformers import pipeline
whisper = pipeline('automatic-speech-recognition', model = 'openai/whisper-medium', device = 0) # alternatively you can use 'openai/whisper-large' if you have moreo GPU memory

# you can add an arguments to choose a specific language, for example hebrew: 
#whisper = pipeline('automatic-speech-recognition', model = 'openai/whisper-large', lang = 'he', device = 0)
```

```python
# function to convert to srt file
from datetime import timedelta
import os
import whisper

def transcribe_audio(path, dest_path):
    model = whisper.load_model("base") # Change this to your desired model
    print("Whisper model loaded.")
    transcribe = model.transcribe(audio=path)
    segments = transcribe['segments']

    for segment in segments:
        startTime = str(0)+str(timedelta(seconds=int(segment['start'])))+',000'
        endTime = str(0)+str(timedelta(seconds=int(segment['end'])))+',000'
        text = segment['text']
        segmentId = segment['id']+1
        segment = f"{segmentId}\n{startTime} --> {endTime}\n{text[1:] if text[0] is ' ' else text}\n\n"

        srtFilename = os.path.join("SrtFiles", dest_path)
        with open(srtFilename, 'a', encoding='utf-8') as srtFile:
            srtFile.write(segment)

    return srtFilename
```

```python
# transcribe the audio
audio_source_path = '/content/drive/MyDrive/71254_2023/01_Lectures/Class10/lecture/intro.mp3'
srt_dest_path = '/content/drive/MyDrive/71254_2023/01_Lectures/Class10/lecture/intro.srt'
transcribe_audio(audio_source_path, srt_dest_path) # call the function, to transcribe to text (to srt file)
```

## Clean a srt file 

remove timestamps, line numbers, and line breaks.

```python
import os

def clean_srt(srt_file):
  # Open the srt file and read its contents
  with open(srt_file, 'r') as f:
    contents = f.read()

  # Split the contents into lines
  lines = contents.split('\n')

  # Remove the timestamps and line numbers from the lines
  cleaned_lines = []
  for line in lines:
    if not line.isdigit() and '-->' not in line:
      cleaned_lines.append(line)

  # Join the cleaned lines into a single string, removing all line breaks
  cleaned_contents = ' '.join(cleaned_lines)

  # Get the base name of the srt file (without the extension)
  base_name = os.path.splitext(srt_file)[0]

  # Create the name of the output txt file
  txt_file = f'{base_name}_cleaned2.txt'

  # Write the cleaned contents to the txt file
  with open(txt_file, 'w') as f:
    f.write(cleaned_contents)
```
